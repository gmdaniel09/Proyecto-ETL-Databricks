```sql
%python
dbutils.widgets.removeAll()
```

```sql
%python
dbutils.widgets.text("catalogo", "catalog_gamd")
catalogo = dbutils.widgets.get("catalogo")
```

```sql
GRANT USE SCHEMA ON SCHEMA ${catalogo}.bronze TO `Desarrollo`;
GRANT CREATE ON SCHEMA ${catalogo}.bronze TO `Desarrollo`;
```

```sql
GRANT USE SCHEMA ON SCHEMA ${catalogo}.silver TO `Desarrollo`;
GRANT CREATE ON SCHEMA ${catalogo}.silver TO `Desarrollo`;
```

```sql
GRANT READ FILES
ON EXTERNAL LOCATION `exlt-raw`
TO `Desarrollo`;
```

```sql
GRANT ALL PRIVILEGES
ON EXTERNAL LOCATION `exlt-bronze`
TO `Desarrollo`;
```
