###INGESTA BRONZE

#### Origen de datos aisles

```sql
%python

# ruta de fuente CSV
ruta = f"abfss://{container}@{storageName}.dfs.core.windows.net/aisles.csv"

# Leer el archivo CSV y guardar en dataframe
df_aisles = spark.read.format("csv")\
.option('header', True)\
.option('sep', ";")\
.option("inferSchema", "true")\
.load(ruta)
```

```sql
%python
df_aisles.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.aisles")
```

### Origen de datos Azure SQL DB:

Tablas: users, deparments y products

```sql
%python
import getpass
```

```sql
%python
jdbcHostname = "sqlgamd.database.windows.net"
jdbcDatabase = "sqlstore"
jdbcPort = 1433
jdbcUsername = "sa_gamd"
jdbcPassword = getpass.getpass("Introduce la contraseña de la base de datos: ")
```

```sql
%python
# URL JDBC
jdbcUrl = f"jdbc:sqlserver://{jdbcHostname}:{jdbcPort};database={jdbcDatabase};encrypt=true;trustServerCertificate=false;hostNameInCertificate=*.database.windows.net;loginTimeout=30;"
```

### Table Deparments

```sql
%python
tableName = "departments"

df_departments = spark.read.format("jdbc") \
    .option("url", jdbcUrl) \
    .option("dbtable", tableName) \
    .option("user", jdbcUsername) \
    .option("password", jdbcPassword) \
    .load()
```

```sql
%python
df_departments.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.departments")
```
