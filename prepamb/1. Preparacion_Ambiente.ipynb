# Preparación de ambientes

1.- Crear Widgets.

2.- Crear external location para las capas bronce, silver y golden.
  - External location: extl-bronze, extl-silver y extl-golden

3.- Crear catalogos y esquemas
  - Catalog: catalog_gamd
  - Esquemas: bronze, silver y golden

## 1.- Widgets

```sql
%python
dbutils.widgets.removeAll()
```

```sql
%python
dbutils.widgets.text("storageName", "adlsgamd26")
```

```sql
%python
storageName = dbutils.widgets.get("storageName")
#print(storageName)
```

## 2.- External Location

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-data`
URL 'abfss://data@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential_v2)
COMMENT 'Ubicación externa para las tablas unit-catalog del Data Lake';
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-raw`
URL 'abfss://raw@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential_v2)
COMMENT 'Ubicación externa para las tablas raw del Data Lake';
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-bronze`
URL 'abfss://bronze@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential_v2)
COMMENT 'Ubicación externa para la capa bronze en el Data Lake';
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-silver`
URL 'abfss://silver@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential_v2)
COMMENT 'Ubicación externa para la capa silver en el Data Lake';
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-golden`
URL 'abfss://golden@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential_v2)
COMMENT 'Ubicación externa para la capa golden en el Data Lake';
```

## 3.- Esquemas y catalogos

```sql
DROP CATALOG IF EXISTS catalog_gamd CASCADE;
```

```sql
CREATE CATALOG IF NOT EXISTS catalog_gamd
MANAGED LOCATION 'abfss://data@${storageName}.dfs.core.windows.net/'
COMMENT 'Catalogo para la arquitectura medallion del ambiente de dev';
```

```sql
CREATE SCHEMA IF NOT EXISTS catalog_gamd.bronze;
CREATE SCHEMA IF NOT EXISTS catalog_gamd.silver;
CREATE SCHEMA IF NOT EXISTS catalog_gamd.golden;
```
